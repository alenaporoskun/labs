###### Теорія

Система динамічних рівнянь
$$ 
\left\{
\begin{array}{rcl}
\frac{dy}{dt}=f(x,y,b,r) \\ \frac{dx}{dt}=g(x,y,b,r) 
\end{array}
\right.
$$

Стаціонарні стани
$$ f(x,y) = 0  \ \ g(x,y) = 0 $$

Стійкість стаціонарних станів

Матриця Якобі
$$
M = \left( \begin{array}{ccc}
\frac{df(x,y)}{dx}|_{x_0,y_0} - \lambda &  \frac{df(x,y)}{dy}|_{x_0,y_0} \\ 
\frac{dg(x,y)}{dx}|_{x_0,y_0}           &  \frac{dg(x,y)}{dy}|_{x_0,y_0} - \lambda \\
\end{array} \right)
\\ det(M) = 0 \rightarrow \lambda 
$$

Характеристичне рівняння і типи стаціонарних точок
$$ A\lambda^2 + B\lambda + C = 0   $$

* $ \tau_{x}=1, \sigma=1: \\ $
$ \tau_{x} \frac{dx}{dt} = \sigma(y-x) = y-x = g(x,y,b,r)  \\ $

*Тепер складемо*  ***матрицю Якобі***, *щоб проаналізувати стійкість стаціонарних станів.*

$$
M = \left( \begin{array}{ccc}
\frac{df(x,y)}{dx}|_{x_0,y_0} - \lambda &  \frac{df(x,y)}{dy}|_{x_0,y_0} \\ 
\frac{dg(x,y)}{dx}|_{x_0,y_0}           &  \frac{dg(x,y)}{dy}|_{x_0,y_0} - \lambda \\

\\ det(M) = 0 \rightarrow \lambda

$
A_{11} = \frac{df(x,y)}{dx}|_{x_0,y_0} - \lambda = (r - \frac{2y}{b}x)|_{x_0,y_0} - \lambda \\
A_{12} = \frac{df(x,y)}{dy}|_{x_0,y_0} = (-\frac{x^2}{b}-1)|_{x_0,y_0} \\
A_{21} = \frac{dg(x,y)}{dx}|_{x_0,y_0} = -1 \\
A_{22} = \frac{dg(x,y)}{dy}|_{x_0,y_0} = 1 - \lambda \\
$

$
det(M) =
\begin{vmatrix} (r-\frac{2y}{b}x)|_{x_0,y_0}-\lambda & (-\frac{x^2}{b}-1)|_{x_0,y_0} \\ -1 & 1-\lambda \end{vmatrix} \  
\rightarrow \ [Т.як \ y=x] \ \rightarrow \
\begin{vmatrix} (r-\frac{2x^2}{b})|_{x_0,y_0}-\lambda & (-\frac{x^2}{b}-1)|_{x_0,y_0} \\ -1 & 1-\lambda \end{vmatrix}
$

$$
\Delta = det(M) = 
\begin{vmatrix} (r-\frac{2x^2}{b})|_{x_0,y_0}-\lambda & (-\frac{x^2}{b}-1)|_{x_0,y_0} \\ -1 & 1-\lambda \end{vmatrix}
$$

 (r-\lambda)(1-\lambda) - (-1)(-1) = r-r\lambda-\lambda+\lambda^2-1 = \\ = \lambda^2 - (r+1)\lambda + (r-1) \\

Далі побудуємо фазові портрети на основі системи динамічних рівнянь: 
$ \\
\left\{
\begin{array}{rcl}
\frac{dy}{dt}= xr - \frac{x^2y}{b}-y = f(x,y,b,r) \\ \frac{dx}{dt}= y-x = g(x,y,b,r) 
\end{array}
\right.
$

```python
def dF1(y, x, r, b):
    return x*r-(x**2 * y)/b - y,  y-x

def fun_plotPhasePortrait(r, b, colors='m'):
    #print("b ",b, "\nr ", r, "\n")
    Y, X = np.meshgrid(np.linspace(-3.0, 3.0, 30), np.linspace(-3.0, 3.0, 30))
    U, V = dF1(Y, X, r, b)
    #plt.streamplot(Y, X, U, V, color=colors, linewidth=0.5, density=1.6)
    plt.streamplot(Y, X, V, U, color=colors, linewidth=0.5, density=1.6)
    j=3
    plt.hlines(0, -j, j, 'k', linewidth=0.5) # label='y=0'
    plt.vlines(0, -j, j, 'k', linewidth=0.9) # label='x=0'
    #plt.legend(bbox_to_anchor=(1,1), loc="lower left") #loc="upper left"
    plt.title("Графік  при r = " f'{r}' ", b = " f'{b}')
    plt.axis('square')
    plt.axis([-j, j, -j, j])
    plt.xlabel("x")
    plt.ylabel("dx/dt")
    #plt.grid()
    #plt.show()

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
fun_plotPhasePortrait(r=0.1, b=0.5, colors='r')       # color = red
plt.subplot(2,2,2)
fun_plotPhasePortrait(r=0.1, b=5, colors=[1, 0.5, 0]) # color = orange
plt.subplot(2,2,3) 
fun_plotPhasePortrait(r=10, b=5, colors='b')          # color = blue
plt.subplot(2,2,4)
fun_plotPhasePortrait(r=10, b=10)
```
