###### Стаціонарні стани 

Динамічне рівняння

$$ \frac{dx}{dt}=f(x,a,b) $$

Стаціонарні стани

$$ f(x) = 0 $$

Стійкість стаціонарних станів

$$ \lambda = \frac{df}{dx} |_{x=x_i} $$
$$ (\lambda < 0 - стійкий \ стан, \lambda > 0 - нестійкий \ стан) $$

**Р-ня**

Знайдемо динамічне рівняння $\frac{dx}{dt}$.

* $ \tau_{z}=0: \\ $
$ \tau_{z} \frac{dz}{dt} = xy-bz = 0 \\ z = \frac{xy}{b} \\ $

* $ \tau_{y}=0: \\ $
$ \tau_{y} \frac{dy}{dt} = x(r-z)-y = x(r-(\frac{xy}{b}))-y = xr - \frac{x^2y}{b}-y = 0  \\
  y(1 + \frac{x^2}{b}) = xr \\ y = \frac{xr}{1+\frac{x^2}{b}} = \frac{bxr}{b+x^2} $
  
* $ \tau_{x}: \\ $
$ \tau_{x} \frac{dx}{dt} = \sigma(y-x) = \frac{bxr}{b+x^2} - x = \frac{bxr - xb - x^3}{b+x^2} = \frac{x(br-b-x^2)}{b+x^2} \\$
Так як $\tau_{x} = 1$, отримуємо динам. рівняння: 
$$ \frac{dx}{dt} = \frac{x(br - b - x^2)}{b+x^2} = f(x, r, b) $$

Розглянемо стійкість стаціонарних станів.
$$\\ \lambda = \frac{df}{dx} |_{x=x_i} $$

$$ f(x, r, b) = \frac{x(br - b - x^2)}{b+x^2} = \frac{bxr - xb - x^3}{b+x^2} $$

$$ 
\frac{df}{dx} = \frac{(br-b-3x^2)(b+x^2)-2x(bxr - xb - x^3)}{(b+x^2)^2} = \frac{b^2r-b^2-2bx^2-brx^2-x^4}{(b+x^2)^2} = 
\frac{(b^2r-brx^2)-(b^2+2bx^2+(x^2)^2)}{(b+x^2)^2} = \frac{br(b-x^2)-(b+x^2)^2}{(b+x^2)^2} = \frac{br(b-x^2)}{(b+x^2)^2}-1 
$$
Отже, 
$$\\ \frac{df}{dx} = \frac{br(b-x^2)}{(b+x^2)^2}-1 $$
$$\\ \lambda = \frac{df}{dx} |_{x=x_i} $$

Обчислимо значення $t$ за формулою отриманою раніше з інтегруванння динамічного рівняння. На виході отримаємо складний список, що складається з значень $x,  r,  b, t$.

```python
def fun_calcD_r1(r, b):
    #      x,  r,  b, t
    D = [ [], [], [], [] ]
    for ri in r:
        for bi in b:
            xi = 0
            t_ = bi*ri*math.log(bi + xi**2)/2 - (xi**2)/2
            t_ = round(t_,3)
            D[0].append(xi)
            D[1].append(ri)
            D[2].append(bi)
            D[3].append(t_)       
    print("len(D[3]) =", len(D[3]))
    return D

D = fun_calcD_r1(r, b)
print("D = array", np.shape(D))
print("D = [ [x],  [r],  [b],  [t] ]")
```

Обчислимо значення $t$ за формулою отриманою раніше з інтегруванння динамічного рівняння. На виході отримаємо складний список, що складається з значень $x_1, x_2, x_3, r,  b, t$.

```python
def fun_calcD_r2(r, b):
    #     x1, x2, x3, r,  b,  t1, t2, t3
    D = [ [], [], [], [], [], [], [], [] ]
    #      0  1   2   3   4   5   6   7
    #print("D = ", D)
    #print("x1i x2i x3i ri bi t_1, t_2, t_3")
    for ri in r:
        for bi in b:
            xi_1 = 0
            t_1 = bi*ri*math.log(bi + xi_1**2)/2 - (xi_1**2)/2
            t_1 = round(t_1,3)
            
            xi_2 = (bi*(ri-1))**(0.5)
            xi_2 = round(xi_2, 3)
            t_2 = bi*ri*math.log(bi + xi_2**2)/2 - (xi_2**2)/2
            t_2 = round(t_2,3)
            
            xi_3 = -xi_2
            t_3 = bi*ri*math.log(bi + xi_3**2)/2 - (xi_3**2)/2
            t_3 = round(t_3,3)
            
            D[0].append(xi_1)
            D[1].append(xi_2)
            D[2].append(xi_3)
            D[3].append(ri)
            D[4].append(bi)
            D[5].append(t_1)
            D[6].append(t_2)
            D[7].append(t_3)
    
    print("len(D[5]) =", len(D[5]))
    return D

D2 = fun_calcD_r2(r2, b2)
print("D2 = array", np.shape(D2))
print("D2 = [ [x1], [x2], [x3], [r], [b], [t1], [t2], [t3] ]")
```
